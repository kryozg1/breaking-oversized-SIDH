```python
#run to define functions
import multiprocessing
from sage.all import factor

def factorize(n, queue):
    try:
        result = factor(n)
        queue.put(result)
    except Exception as e:
        queue.put(f"Error: {e}")

def safe_factor(n, timeout=5):
    queue = multiprocessing.Queue()
    process = multiprocessing.Process(target=factorize, args=(n, queue))
    
    process.start()
    process.join(timeout)

    if process.is_alive():
        process.terminate()
        process.join()
        return f"Timeout for {n}"
    
    return queue.get()

def OddCyclicSumOfSquares(n, factexpl):
    fac = factexpl
    if not set([f % 4 for f in fac]) == set([1]):
        return False, 0, 0
    else:
        Z = ZZ[I]
        prod = 1
        for f in fac:
            p = f
            F = GF(p)
            z = F.random_element()
            while not z^((p-1)//2) == F(-1):
                z = F.random_element()
            z = z^((p-1)//4)
            z = int(z)
            prod *= gcd(Z(p), Z(z+I))
        u = [int(e) for e in prod]
        if u[0]%2==1: 
            u1 = u[0]
            u2 = u[1]//2 
        else:
            u1 = u[1]
            u2 = u[0]//2
        return True, abs(u1), abs(u2)
```

```python
#run to find pairs (b_i, a_i)
def TD(x, pt):
    r = x
    for p in pt:
        i = 0
        while r % p == 0:
            if p%4 == 3:
                print(f'p equiv 3 mod 4 (!)')
                return False, 1
            r //= p
            i += 1
            if i > 1:
                print(f'exponent > 1 (!)')
                return False, r
        if r == 1:
            return True, 1
    return True, r

a = 382 
b = 241 #Change to desired value
while True:
    ea = a
    eb = b
    nf = True
    while nf:
        c = 2^ea-3^eb
        if c < 0:
            ea += 1
            continue
        print(f'b:{b}-a:{ea}.', end='')
        d, r = TD(c, pt)
        if not d:
            ea += 1
            continue
        if r%4 == 3:
            print('reminder equiv 3 mod 4 (!)')
            ea += 1
            continue
        print(f'may be correct (?)')
        f = safe_factor(r, timeout=600)
        print(f'\tsafe_factor result: {f if not f[0] == "T" else "Timeout (!)"}')
        if f[0] == "T":
            ea += 1
            continue
        else:
            if True in [fac[0]%4==3 or fac[1] > 1 for fac in f]:
                print(f'reminder divisor equiv 3 mod 4 or with exponent > 1 (!)')
                ea += 1
                continue
            else:
                nf = False
    b += 2
    while 3^b > 2^a:
        a += 1
```
